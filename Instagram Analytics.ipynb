```python
plt.figure(figsize= (15,6))
plt.bar(df['content_category'].sort_values(), df['impressions'].sort_values(), color= 'green')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
hourly = df.groupby(['post_hour', 'content_category'])['impressions'].median().unstack()

fig, ax = plt.subplots(figsize=(16, 5))

n_categories = len(hourly.columns)
width = 0.8 / n_categories
x = np.arange(len(hourly.index))

for i, category in enumerate(hourly.columns):
    ax.bar(
        x + i * width,
        hourly[category],
        width=width,
        label=category,
        edgecolor='none'
    )

ax.set_xlabel('Hour of Day')
ax.set_ylabel('Avg Impressions')
ax.set_xticks(x + width * (n_categories - 1) / 2)
ax.set_xticklabels(hourly.index)
ax.legend(title='Content Category')


plt.tight_layout()
plt.show()
```
